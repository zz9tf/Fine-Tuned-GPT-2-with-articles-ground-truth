```python
import os
import sys
sys.path.insert(0, '../../..')
from component.io import load_nodes_jsonl

database_node_file_path = os.path.abspath('../../../database/gpt-4o-batch-all-target/0.jsonl')
nodes = load_nodes_jsonl(database_node_file_path)
node_id_to_node = {node.id_:node for node in nodes}
```

```python
import pandas as pd

df = pd.read_csv('../SciFive-base-PMC_results/SciFive-base-PMC_ids_labels_preds.csv')
```

```python
total = 0
lower_than_1 = 0
lower_than_1_nodes_labels = []
for index, row in df.iterrows():
    if row['True Label'] > 1 and row['True Label'] != row['Prediction']:
        total += 1
        if row['Prediction'] <= 1:
            lower_than_1 += 1
            lower_than_1_nodes_labels.append((node_id_to_node[row['ID']], row['Prediction'], row['True Label']))
            
```

```python
import pandas as pd

# Helper function to find the position of a title in a list
def find_position(title_list, v):
    for i, x in enumerate(title_list):
        if x == v:
            return i
    return -1

num_to_level = {i:level for i, level in enumerate(['document', 'section', 'paragraph', 'multi-sentences'])}

# List to store data for each node
data = []

# Assuming `lower_than_1_nodes_labels` and `node_id_to_node` are defined
for (node, prediction, actual) in lower_than_1_nodes_labels:
    node_id = node.id_
    predicted_label = num_to_level[prediction]
    actual_label = num_to_level[actual]
    
    # Traverse up to the 'section' level in the hierarchy
    while node.parent_node.metadata['level'] != 'section':
        node = node_id_to_node[node.parent_node.node_id]
    
    parent_id = node.parent_node.node_id
    parent_node = node_id_to_node[parent_id]
    parent_title = parent_node.metadata['section_title']
    
    # Find the section position
    section_position = find_position(
        list(parent_node.parent_node.metadata['sections'].keys()), 
        parent_title
    )
    
    # Append data as a dictionary
    data.append({
        "node_id": node_id,
        "predicted_label": predicted_label,
        "actual_label": actual_label,
        "section_id": parent_id,
        "section_title": parent_title,
        "section_position": section_position
    })

# Create DataFrame from the list of dictionaries
track_df = pd.DataFrame(data)
```
